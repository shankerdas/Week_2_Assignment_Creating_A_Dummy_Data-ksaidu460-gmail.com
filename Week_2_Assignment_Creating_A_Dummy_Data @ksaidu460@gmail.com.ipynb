```python
import numpy as np
mean1 = (-3,0)#this is the mean

#This is the covariance matrix
cov = [[1.0,0.8],[0.8,1.0]]

#To ensure that the value does not change from one run to the next
np.random.seed(0)

#The shape of the created ndarray will be (500, 2)
random_ndarray = np.random.multivariate_normal(mean1,cov,size=500)
print(random_ndarray.shape)
print(random_ndarray)
```

### Problem 2 Visualization by scatter plot

```python
#we use matplotlib plt.scatter() to visualize it by scatter plot
import matplotlib.pyplot as plt
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(random_ndarray[:,0], random_ndarray[:,1], color = "blue",alpha=0.7)
plt.show()
```

### Problem 3 Visualization by histogram

```python
import matplotlib.pyplot as plt
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histogram")
plt.hist(random_ndarray[:,0], color = "orange")
plt.xlim(left=-4)#This determin the range of the horizontal axis for left
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("histogram of x2")
plt.hist(random_ndarray[:,1], color = "orange")
plt.xlim(left=-5)#This determin the range of the horizontal axis for right
plt.show()
```

### Problem 4 Addition of data

```python
mean2 = (0,-3)
cov = [[1.0,0.8],[0.8,1.0]]
np.random.seed(0)
second_random_ndarray = np.random.multivariate_normal(mean2,cov,size=500)

#we now visualize the data of problems 1 and 4 as one scatter plot
plt.xlabel("x1", color = "blue")
plt.ylabel("x2", color = "blue")
plt.title("scatter", color = "orange")

plt.scatter(random_ndarray[:, 0], random_ndarray[:, 1], label="0")
plt.scatter(second_random_ndarray[:, 0], second_random_ndarray[:, 1], label="1")
plt.legend()
plt.show()
```

### Problem 5 Data combination

```python
np.random.seed(0)

#This code combines the data in problem 1 and 4
combined_ndarray = np.concatenate((random_ndarray,second_random_ndarray))
print(combined_ndarray.shape)
print(combined_ndarray)
```
